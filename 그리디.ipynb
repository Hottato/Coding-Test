```python
n=25;k=5
result = 0

while True:
    target = (n//k)*5
    result+=(n-target)
    if n <k:
        break
    result+=1
    n//=k

result += (n-1) 
# 만약 n이 3이다 -> 그러면 3-1=2 이뜻은 뺴기 연산을 2번 더해야 한다는 뜻과 같다
print(result)
```

# 구현

```python
n= int(input())
```

```python
control = input().split()
```

```python
control
```

```python
x,y= (1,1)
```

```python
n
```

```python
dx = [1,-1,0,0] #오+,왼-

dy = [0,0,-1,1] # 위 -1, 아 1
```

```python
while n == 0:
    for i in control:
        if i == 'R':
            x+=dx[0]
            n-=1
            if x==0:
                x+=1
        elif i == 'L':
            x+=dx[1]
            n-=1
            if x==0:
                x+=1  
            
        elif i == 'U':
            y+=dy[2]
            n-=1
            if y==0:
                y+=1 
                
        elif i == 'D':
            y+=dy[3]
            n-=1
            if y==0:
                y+=1 
                
print(x,y)
```
