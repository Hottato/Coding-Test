`-` 순열과 조합

````python
from itertools import combinations
````

````python
list(combinations(data,1))
````

````python
from itertools import permutations,combinations,combinations_with_replacement,product


````

````python
data = ["a","b","c"]

result = list(permutations(data,3)) #모든 순열 구하기

print(result)



````

````python
result_com = list(combinations(data,2)) # 2개를 뽑는 모든 조합
print(result_com)
````

````python
result_comre = list(combinations_with_replacement(data,2)) # 조합 중복허용
print(result_comre)
````

````python
result = list(product(data,repeat = 2)) # 중복 허용 순열
print(result)
````

````python
input_string = input("문자열을 입력하세요: ")
uppercase_letters = []

for char in input_string:
    if char.isupper():
        uppercase_letters.append(char)

result = ''.join(uppercase_letters)
print("출력 결과:", result)
````
